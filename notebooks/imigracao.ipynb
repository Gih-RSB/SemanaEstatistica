## Transformando um grafico em figura

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Definir o estilo do gráfico usando seaborn
sns.set(style="whitegrid")

# Configuração da figura
plt.figure(figsize=(12, 6))

# Plotagem dos dados com linha sólida e marcadores
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'], 
         color='darkblue', marker='o', linestyle='-', linewidth=2, markersize=8)

# Adiciona uma área de preenchimento abaixo da linha
plt.fill_between(dados_brasil['ano'], dados_brasil['imigrantes'], color='lightblue', alpha=0.4)

# Configuração dos ticks no eixo x
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'], rotation=45)

# Configuração dos ticks no eixo y
plt.yticks(range(0, max(dados_brasil['imigrantes']) + 1, 1000))

# Adiciona a grade ao gráfico
plt.grid(True, linestyle='--', alpha=0.7)

# Adiciona o título do gráfico
plt.title('Imigração do Brasil para o Canadá', fontsize=16, fontweight='bold')

# Adiciona rótulos aos eixos
plt.xlabel('Ano', fontsize=14, fontweight='bold')
plt.ylabel('Número de Imigrantes', fontsize=14, fontweight='bold')

# Adiciona uma legenda
plt.legend(['Número de Imigrantes'], loc='upper left', fontsize=12)

# Exibe o gráfico
plt.show()

```

```python

sns.set(style="whitegrid")
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

# Subplot 1: Gráfico de Linha
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'], color='dodgerblue', marker='o', linestyle='-', linewidth=2, markersize=8)
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: str(int(x))))  # Formata o eixo x para mostrar os anos como inteiros
axs[0].set_xlabel('Ano', fontsize=12, fontweight='bold')
axs[0].set_ylabel('Número de Imigrantes', fontsize=12, fontweight='bold')
axs[0].set_title('Número de Imigrantes ao Longo dos Anos', fontsize=14, fontweight='bold')
axs[0].grid(True, linestyle='--', alpha=0.7)

# Subplot 2: Gráfico Boxplot
axs[1].boxplot(dados_brasil['imigrantes'], patch_artist=True, 
               boxprops=dict(facecolor='lightcoral', color='darkred'), 
               whiskerprops=dict(color='darkred'),
               capprops=dict(color='darkred'),
               medianprops=dict(color='darkred'),
               flierprops=dict(markerfacecolor='red', marker='o'))
axs[1].set_title('Boxplot da Imigração do Brasil \n para o Canadá (1980-2013)', fontsize=14, fontweight='bold', color='darkorange')
axs[1].set_xlabel('Brasil', fontsize=12, fontweight='bold', color='blue')
axs[1].set_ylabel('Número de Imigrantes', fontsize=12, fontweight='bold', color='green')


plt.tight_layout()

plt.show()
```

## Comparativo de imigracoes do brasil com alguns países da america latina

```python



fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.9, wspace=0.5)

axs[0, 0].plot(df.index, df['Brasil'], marker='o', color='blue')
axs[0, 0].set_title('Brasil')
axs[0, 0].set_xlabel('Ano')
axs[0, 0].set_ylabel('Valor')
axs[0, 0].grid(True)

axs[0, 1].plot(df.index, df['Colombia'], marker='o', color='green')
axs[0, 1].set_title('Colombia')
axs[0, 1].set_xlabel('Ano')
axs[0, 1].set_ylabel('Valor')
axs[0, 1].grid(True)

axs[1, 0].plot(df.index, df['Argentina'], marker='o', color='red')
axs[1, 0].set_title('Argentina')
axs[1, 0].set_xlabel('Ano')
axs[1, 0].set_ylabel('Valor')
axs[1, 0].grid(True)

axs[1, 1].plot(df.index, df['Peru'], marker='o', color='orange')
axs[1, 1].set_title('Peru')
axs[1, 1].set_xlabel('Ano')
axs[1, 1].set_ylabel('Valor')
axs[1, 1].grid(True)

plt.tight_layout()
plt.show()






```

```python
# criar uma figura
fig, axs = plt.subplots(2,2,figsize=(10,6))
# wspace - Espaço Horizontal 
# hspace - Espaço Verticaç
fig.subplots_adjust(hspace=0.9,wspace=0.5)
# subplots
# Brasil
axs[0,0].plot(df.loc['Brasil',anos] ,color = 'green')
axs[0,0].set_title('Brasíl')

# Colômbia
axs[0,1].plot(df.loc['Colômbia',anos],color = 'yellow')
axs[0,1].set_title('Colômbia')

# Argentina
axs[1,0].plot(df.loc['Argentina',anos])
axs[1,0].set_title('Argentina')

# Peru
axs[1,1].plot(df.loc['Peru',anos],color = 'red')
axs[1,1].set_title('Peru')

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid(True)
```
